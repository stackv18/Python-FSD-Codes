```python
%%writefile String_Fun.py
def Func(user_input):
    print("Lenght is:",len(user_input))
    print("Upper is:",user_input.upper())
    print("Lower is:",user_input.lower())
    print("Capitalize is:",user_input.capitalize())
    print("Is upper? :",user_input.isupper())
    print("Is lower? :",user_input.islower())
```
