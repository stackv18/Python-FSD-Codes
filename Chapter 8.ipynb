```python
#CUSTOM EXCEPTION
class Mahi(Exception):
    def __init__(self,msg):
        self.msg=msg

def add(a,b):
    if type(a)==str:
        raise Mahi("a str ma na chale nikal")
    if type(b)==str:
        raise Mahi("b str ma na chale hed hed")
    return(a+b)
add("abc",12)
```

```python
#PRINT DETAIL FUNCTION
class Employee:
    def __init__(self,name,age,salary,increament):
        self.name=name
        self.age=age
        self.salary=salary
        self.increament=increament
        self.printdetail() #cool
    def printdetail(self):
        print(f"{self.name} has salary {self.salary} and is {self.age} years old")
        old=self.salary
        salary=self.salary*self.increament
        print(f"Salary after increament: {self.name}: {old} ---> {salary} \n")

#manually printing
#obj0=Employee("Narendra",400) ERROR Not enough args
obj1=Employee("Manish",30,40000,1.3)
obj2=Employee("Suresh",69,1000011,1.7)
obj3=Employee(age=30,salary=203000,name="Ram",increament=1.6)

print(obj1.salary)
print(f"\nOBJ1 \nname:{obj1.name} \nage:{obj1.age} \nsalary:{obj1.salary}")
print(f"\nOBJ2 \nname:{obj2.name} \nage:{obj2.age} \nsalary:{obj2.salary}")
print(f"\nOBJ3 \nname:{obj3.name} \nage:{obj3.age} \nsalary:{obj3.salary}")

print("\n")

obj1.printdetail()
```

```python
def fun():
    return 10
    return 20
    return 30
x=fun()
print(x)
#for i in x:
#    print(i)
#ERROR because of return
#TypeError: 'int' object is not iterable
```

```python
def fun():
    yield 10
    yield 20
    yield 30
x=fun()
print(x)
for i in x:
    print(i)
```
