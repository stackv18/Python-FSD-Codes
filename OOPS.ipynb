```python
class String:
    def __init__(self):
        self.vowel=0
        self.consonent=0
        self.uppercase=0
        self.lowercase=0
        self.space=0
        self.string=input("Enter String")
    def count_uppercase(self):
        for i in self.string:
            if i.isupper():
                self.uppercase+=1
    def count_lowercase(self):
        for i in self.string:
            if i.islower():
                self.lowercase+=1
    def count_vowel(self):
        for i in self.string:
            if i in "aeiouAEIOU":
                self.vowel+=1
    def count_space(self):
        for i in self.string:
            if i.isspace():
                self.space+=1
    def count_consonent(self):
        for i in self.string:
            if i.isspace() or i.isdigit():
                pass
            if i not in "aeiouAEIOU":
                self.consonent+=1
    def count_all(self):
        self.count_uppercase()
        self.count_lowercase()
        self.count_vowel()
        self.count_space()
        self.count_consonent()
    def display(self):
        print(self.uppercase)
        print(self.lowercase)
        print(self.space)
        print(self.consonent)
        print(self.vowel)
```

```python
x=String()
x.count_all()
x.display()
```
