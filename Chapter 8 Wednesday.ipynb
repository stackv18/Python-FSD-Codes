```python
#ATM 9 Markers
'''OPTIONS: 1. Create Pin // 2. Change Pin // 3. Checkbalance // 4. Withdrawal // 5. EXIT if any key pressed'''
```

```python
class Atm:
    def __init__(self):
        self.balance=50000
        self.menu()
    def menu(self):
        self.choice=int(input("Enter your choice: \n1. Create Pin \n2. Change Pin \n3. Withdrawl \n4. Check balance \n5. Exit\n"))
        if self.choice == 1:
            self.createpin()
        elif self.choice == 2:
            self.changepin()
        elif self.choice == 3:
            self.validpin()
        elif self.choice == 4:
            self.checkbalance()
        elif self.choice == 5:
            self.exit()
    def createpin(self):
        self.pin=int(input("Enter the Pin: "))
        self.menu()
        
    def changepin(self):
        pin=int(input("Enter the current pin: "))
        if pin==self.pin:
            newpin=int(input("Enter the new pin: "))
            if newpin==self.pin:
                print("New pin must be different from old one")
            else:
                self.pin=newpin
        else:
            print("Enter your correct pin")
        self.menu()
        
    def checkbalance(self):
        print(f"Your current balance is: {self.balance}")
        self.menu()
        
    def withdraw(self):
        
        withdrawal=int(input("Enter the amount of money to withdraw: "))
        if withdrawal < self.balance:
            self.balance = self.balance - withdrawal
            print("Withdrawal Successfull")
        else:
            print("Not enough balance")
        self.menu()
        
    def exit(self):
        return
    
    def validpin(self):
        spin=int(input("Enter your pin: "))
        if spin == self.pin:
            self.withdraw()
        else:
            print("Invalid Pin")

obj=Atm()
```

```python
class student:
    schoolname="LJEIT" #class varaible
    def __init__(self,name,age,schoolname):
        self.name=name   #instance
        self.age=age     #instance
        self.schoolname=schoolname
    def display(self):
        print("This is a private function")
        print(self.age)
        
x=student("Ramesh",32,"SVJK")
print(x.schoolname)
print(student.schoolname)
print(x.name)
print(x.age)
x.display()
x.age=70
x.display()
```

```python
#protected variable with_
class student:
    _schoolname="LJEIT" #class varaible
    def __init__(self,name,age,schoolname):
        self._name=name   #instance
        self._age=age     #instance
        self.schoolname=schoolname
        
x=student("Ramesh",32,"SVJK")
print(x._schoolname)
#print(student.schoolname)
print(x._name)
print(x._age)
```

```python
#strongly private variable with __
class student:
    __schoolname="LJEIT" #class varaible
    def __init__(self,name,age):
        self.__name=name   #instance
        self.__age=age     #instance
    def display(self):
        print("This is a private function")
        print(self.__age)
        
x=student("Ramesh",32)
x.display()
x.__age=70
x.display()
```
