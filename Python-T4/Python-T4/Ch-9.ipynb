## Operator overloading: 

```python
print(1+2)
print("1"+"2")
print("Keval"*3)

print(int.__add__(1,2))
print(str.__add__("1","2"))
print(str.__mul__("Keval",3))
```

```python
#operator overloading for complex numbers:
#Arithmetic operators

class Complex:
    def __init__(self,r,i):
        self.real=r
        self.imaginary=i
    
    def __add__(self,other): #self vs other
        real=self.real+other.real
        ima=self.imaginary+other.imaginary
        return f"{real}+{ima}i"
        
c1=Complex(1,4)
c2=Complex(3,5)
print(c1+c2)
```

```python
#Comparision operators

class A:
    def __init__(self,a):
        self.a=a
        
    def __gt__(self,other):
        if self.a>other.a:
            return f"{self.a} is higher"
        else:
            return f"{other.a} is higher"
        
x=A(4)
y=A(3)
print(x>y)
```

## Method overloading: 

```python
class Addition:
    def sum(self,a,b):
        return a+b
    def sum(self,a,b,c):
        return a+b+c
    
obj=Addition()
obj.sum(3,4)
obj.sum(3,4,7)
```

```python
class Addition:
    def sum(self,a,b):
        return a+b
    def sum(self,a,b,c=5):
        return a+b+c
    
obj=Addition()
print(obj.sum(3,4))
print(obj.sum(3,4,7))
```

```python
class Addition:
    def sum(self,a,b):
        return a+b
    def sum(self,*args):
        total=0
        for i in args:
            total+=i
        print(total)
    
obj=Addition()
obj.sum(3,4,8,6)
obj.sum(3,4,7)
```

### Single inheritance: 

```python
class parent:
    def sleep(self):
        print("Sleep at 9 PM")
        
class child:
    def sleep(self):
        print("Sleep at 12 AM")
    def eat(self):
        print("eat only pizzas")
        
obj=parent()
obj.sleep()
```

```python
class parent:
    def sleep(self):
        print("Sleep at 9 PM")
        
class child(parent):
    def sleep(self):
        print("Sleep at 12 AM")
        super().sleep()
    def eat(self):
        print("eat only pizzas")
        
obj=child()
obj.sleep()
```

```python
class parent:
    def sleep(self):
        print("Sleep at 9 PM")
        
class child(parent):
    def swim(self):
        print("swim at 12 AM")
    def eat(self):
        print("eat only pizzas")
        
obj=child()
obj.sleep()
print(parent.mro())
```

### Multiple inheritance: 

```python
class Father:
    def sleep(self):
        print("Sleep at 9 PM")
    def swim(self):
        print("swim at 6 AM")
        
class Mother:
    def cook(self):
        print("cook very good")
    def sleep(self):
        print("sleep at 10 PM")

class Child(Father,Mother):
    def run(self):
        print("run very fast")
        
x=Child()
x.run()
x.sleep()
print(Child.mro())
```

```python
class Father:
    def sleep(self):
        print("Sleep at 9 PM")
    def swim(self):
        print("swim at 6 AM")
        
class Mother:
    def cook(self):
        print("cook very good")
    def sleep(self):
        print("sleep at 10 PM")

class Child(Mother,Father):
    def run(self):
        print("run very fast")
        
x=Child()
x.run()
x.sleep()
print(Child.mro())
```

### Multilevel inheritance: 

```python
class Grandfather:
    def sleep(self):
        print("Sleep at 9 PM")
    def swim(self):
        print("swim at 6 AM")
        
class Father(Grandfather):
    def cook(self):
        print("cook very good")
    def sleep(self):
        print("sleep at 10 PM")

class Child(Father):
    def run(self):
        print("run very fast")
        
x=Child()
x.run()
x.sleep()
print(Child.mro())
```

### Hierarchical inheritance: 

```python
class Parent:
    def display(self):
        print("display from parent class.")
        
class Child1(Parent):
        def fun(self):
            print("Fun from child class.")
            
class Child2(Parent):
        def show(self):
            print("show from child2 class.")
            
obj=Child1()
obj.display()
obj2=Child2()
obj2.show()
print(Child1.mro())
```

### Hybrid inharitance: 

```python
class Parent1:
    def display(self):
        print("display from parent1 class.")
        
class Parent2:
    def cook(self):
        print("cook very good.")
        
class Child(Parent1,Parent2):
    def show(self):
            print("show from child class.")
            
class GC1(Child):
    def fun(self):
            print("Fun from GC1 class.") 
    
class GC2(Child):
    def fun2(self):
            print("Fun from GC2 class.") 
            
obj=GC1()
obj.cook()
obj2=GC2()
obj2.display()
```

### Classic diamond problem (MRO): 

```python
class P1:
    def show(self):
        print("Display from p1.")
        
class P2:
    def fun(self):
        print("Display from p2.")
        
class C1(P1):
    def foo(self):
        print("Display from C1.")
        
class C2(P2):
    def boo(self):
        print("Display from C2.")
        
class GC(C1,C2):
    def display(self):
        print("Display from GC.")
        
obj1=GC()
print(GC.mro())
```

# Abstraction

```python
import abc
dir(abc)
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self,shape_name):
        self.shape_name = shape_name
        
    @abstractmethod
    def draw(self):
        pass
        
class Circle(Shape):
    def __init__(self):
        super().__init__("circle")
    
    def draw(self):
        print("Circle")
        
class Triangle(Shape):
    def __init__(self):
        super().__init__("triangle")
    
    def draw(self):
        print("Triangle")
        
x = Circle()
y = Triangle()
x.draw()
y.draw()
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self,shape_name):
        self.shape_name = shape_name
        
    @abstractmethod
    def draw(self):
        pass
        
class Circle(Shape):
    def __init__(self):
        super().__init__("circle")
    
    def draw1(self):
        print("Circle")
        
        
x = Circle()
x.draw1()
```
