```python
#OPERATOR OVERLOADING for complex number
class Complex:
    def __init__(self,r,i):
        self.real=r
        self.imaginary=i
        
    def __add__(self,other):
        real=self.real+other.real
        ima=self.imaginary+other.imaginary
        return f"{real} + {ima}i"
        
        
c1=Complex(1,4)
c2=Complex(3,5)

c3=c1+c2
print(c3)
```

```python
class A:
    def __init__(self,a):
        self.a=a
    def __gt__(self,other):
        if self.a > other.a:
            return f"{self.a} is bigger"
        else:
            return f"{other.a} is highter"
x=A(4)
y=A(3)
print(x>y)
```

```python
class A:
    def __init__(self,a):
        self.a=a
    def __lt__(self,other):
        if self.a < other.a:
            return f"{self.a} is smaller"
        else:
            return f"{other.a} is lower"
x=A(4)
y=A(3)
print(x<y)
```

```python
#METHOD OVERLOADING
class ADDITION:
    def sum(self,a,b):
        return a+b
    def sum(self,a,b,c=5):
        return a+b+c
obj=ADDITION()
print(obj.sum(2,4))
print(obj.sum(3,4,5))
```

```python
#METHOD OVERLOADING
class ADDITION:
    def sum(self,a,b):
        return a+b
    def sum(self,*args):
        total=0
        for i in args:
            total+=i
        print(total)
        
obj=ADDITION()
obj.sum(2,4)
```

```python
#MULTIPLE INHERITANCE
class Father:
    def sleep(self):
        print("Sleeps at 9pm")
    def swim(self):
        print("Swim at 6am")
class Mother:
    def cook(self):
        print("Cook very good")
    def sleep(self):
        print("Sleeps at 10pm")
class Child(Mother,Father):
    def run(self):
        print("Run like dawg")
x=Child()
x.run()
x.sleep()
print(Child.mro())
```

```python
#MULTILEVEL INHERITANCE
class GrandFather:
    def sleep(self):
        print("Sleeps at 9pm")
    def swim(self):
        print("Swim at 6am")
class Father(GrandFather):
    def cook(self):
        print("Cook very good")
    def sleep(self):
        print("Sleeps at 10pm")
class Child(Father):
    def run(self):
        print("Run like dawg")
x=Child()
x.run()
x.sleep()
x.swim()
print(Child.mro())
#how to access gfather slep from child
```

```python
#HIERARCHICAL INHERITANCE
class Parent:
    def display(self):
        print("display parent")
class Child(Parent):
    def fun(self):
        print("fun at 7 am")
class Child2(Parent):
    def show(self):
        print("show at 7 am")
x=Child2()
y=Child()
y.fun()
x.show()
print(Child.mro())
print(Child2.mro())
```

```python
#HYBRID INHERITANCE
class Father:
    def f(self):
        print("father")
class Mother:
    def f1(self):
        print("mother")
class Child(Father,Mother):
    def display(self):
        print("child")
class dada_no_chokro(Child):
    def gc(self):
        print("grand child 1")
class dada_no_chokro2(Child):
    def gc2(self):
        print("grand child 2")
x=dada_no_chokro()
x.display()
x.f()
x.gc()
x.f1()
print("============")
x=dada_no_chokro2()
x.display()
x.f()
x.gc2()
x.f1()    
print("============")    
print(dada_no_chokro.mro())
```
