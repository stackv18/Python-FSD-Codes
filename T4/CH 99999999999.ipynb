```python
# Operator overloading for complex number
class Complex:
    def __init__(self,r,i):
        self.r=r
        self.i=i
    def __add__(self,other):
        r=self.r+other.r
        i=self.i+other.i
        return f"{r}+{i}i"
        
        
c1=Complex(1,4)
c2=Complex(3,5)
print(c1+c2)
```

```python
class A:
    def __init__(self,a):
        self.a=a
    def __gt__(self,other):
        if self.a>other.a:
            return f"{self.a} is higher"
        else:
            return f"{other.a} is higher"
    def __lt__(self,other):
        if self.a<other.a:
            return f"{self.a} is lower"
        else:
            return f"{other.a} is lower"
x = A(4)
y = A(30)
print(x>y)
print(x<y)
```

# Method overloading

```python
class Addition:
    def sum(self,a,b):
        return a+b
    def sum(self,a,b,c=100):
        return a+b+c
obj=Addition()
print(obj.sum(4,5))
print(obj.sum(4,5,9))
```

```python
class Addition:
    def sum(self,a,b):
        return a+b
    def sum(self,*args):
        return sum(args)
obj=Addition()
obj.sum(10,10,10,10,10)
```

# Method overriding (inheritance)

# Single inheritance

```python
class Parent:
    def sleep(self):
        print("sleep at 9 pm")
class Child(Parent):
    def sleep(self):
        print("sleep at 12 am")
    def eat(self):
        print("eat at 8 am")
x=Parent()
x.sleep()
```

```python
class Parent:
    def sleep(self):
        print("sleep at 9 pm")
class Child(Parent):
    def sleep(self):
        print("sleep at 12 am")
        super().sleep()
    def eat(self):
        print("eat at 8 am")
x=Child()
x.sleep()
x.eat()
print(Parent.mro())
```

```python
class Parent:
    def sleep(self):
        print("sleep at 9 pm")
class Child(Parent):
    def swim(self):
        print("swim at 7 am")
    def eat(self):
        print("eat at 8 am")
x=Child()
x.swim()
x.eat()
print(Child.mro())
```

# Multiple inheritance

```python
class Father:
    def sleep(self):
         print("sleep at 10 pm")
    def swim(self):
        print("swim at 9 pm")
class Mother:
    def sleep(self):
        print("sleep at 10 am")
    def cook(self):
        print("cook very good")
class Child(Father,Mother):
    def run(self):
        print("RUN VERY SLOW")
x = Child()
x.run()
x.sleep()
x.cook()
x.swim()
print(Child.mro())
```

# Multilevel inheritance

```python
class Car:
    def color(self):
        print("black")
class Audi(Car):
    def car_name(self):
        print("name is audi")
class Model(Audi):
    def car_model(self):
        print("q4")
x = Model()
x.color()
x.car_name()
x.car_model()
print(Model.mro())
```

# Hierarchical inheritance

```python
class Parent:
    def display(self):
        print("display parent")
class Child(Parent):
    def fun(self):
        print("fun at 7 am")
class Child2(Parent):
    def show(self):
        print("show at 7 am")
x=Child2()
y=Child()
y.fun()
x.show()
print(Child.mro())
print(Child2.mro())
```

# Hybrid inheritance

```python
class Father:
    def f(self):
        print("father")
class Mother:
    def f1(self):
        print("mother")
class Child(Father,Mother):
    def display(self):
        print("child")
class dada_no_chokro(Child):
    def gc(self):
        print("grand child 1")
class dada_no_chokro2(Child):
    def gc2(self):
        print("grand child 2")
x=dada_no_chokro()
x.display()
x.f()
x.gc()
x.f1()
print("============")
x=dada_no_chokro2()
x.display()
x.f()
x.gc2()
x.f1()    
print("============")    
print(dada_no_chokro.mro())        
```

```python
class p1:
    def f(self):
        print("father")
class p2:
    def f1(self):
        print("mother")
class c1(p1,p2):
    def display(self):
        print("child")
class c2(p2):
    def gc(self):
        print("grand child 1")
class gc(c1,c2):
    def gc2(self):
        print("grand child 2")

print(gc.mro())
```

# Abstraction

```python
import abc
dir(abc)
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self,shape_name):
        self.shape_name = shape_name
        
    @abstractmethod
    def draw(self):
        pass
        
class Circle(Shape):
    def __init__(self):
        super().__init__("circle")
    
    def draw(self):
        print("Circle")
        
class Triangle(Shape):
    def __init__(self):
        super().__init__("triangle")
    
    def draw(self):
        print("Triangle")
        
x = Circle()
y = Triangle()
x.draw()
y.draw()
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self,shape_name):
        self.shape_name = shape_name
        
    @abstractmethod
    def draw(self):
        pass
        
class Circle(Shape):
    def __init__(self):
        super().__init__("circle")
    
    def draw1(self):
        print("Circle")
        
        
x = Circle()
x.draw1()
```
