```python
File Types:
    
    1. TXT files(.txt)
    
    r = read (cursor at begin) (file not found error)
    r+ = read + write(cursor at begin) .write ommits old (file not found error)
    w = write (cursor at begin)
    w+ = write + read (cursor at begin) .write concatinates to old
    a = append (cursor at begin)
    a+ = append + write (cursor at begin)
```

```python
#Write a dictionary data in a text file named output.txt
'''
Name: Divy Bhuta
Age: 32
Degree: ME in structure
Position: Sr. Planner
Firm: Shah & Talati
Branch: Baroda
'''

d={'Name': 'Divy Bhuta', 'Age': '32', 'Degree': 'ME in structure', 'Position': 'Sr. Planner', 'Firm': 'Shah & Talati', 'Branch': 'Baroda'}
f=open('output.txt','w')
for x,y in d.items():
    f.write(str(x)+":"+str(y)+"\n")
    
f.close()
```

```python
f=open("compare1.txt")
f1=open("compare2.txt")

s1=f.read()
s2=f1.read()
line=1
pos=1

for i in range(len(s1)):
    if s1[i]==s2[i]:
        pos+=1
        if s1[i]=="\n":
            line+=1
            pos=1
    else:
        print("Line:",line,"Pos:",pos)
        pos+=1
```
